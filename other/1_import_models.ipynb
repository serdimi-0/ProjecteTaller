```python
import pandas as pd
from pymongo import MongoClient
```

## Connexió

```python
client = MongoClient('mongodb://localhost:27017/')
db = client['taller']
collection = db['models']
```

## Lectura del csv

```python
df = pd.read_csv('idae.csv')
df = df[['Modelo']].rename(columns = {'Modelo':'model'})
df.head()
```

## Traducció a diccionari

```python
dict = df.to_dict(orient='records')
dict[:5]
```

## Introducció a la BD

```python
collection.insert_many(dict)
```
